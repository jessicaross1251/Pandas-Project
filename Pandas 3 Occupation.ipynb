### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'

users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|', index_col='user_id')
users.head(20)
```

### Step 4. Discover what is the mean age per occupation

```python
mean_age_per_occ = users.groupby('occupation')['age'].mean()
print(mean_age_per_occ)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# qty_per_occ = users.groupby('occupation').count()
# male_rat_per_occ = qty_per_occ[(users['gender'] == 'M')]
# print(qty_per_occ)

#create pivot
gender_ratio = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)

#calc male ratio
sums = gender_ratio[['F', 'M']].sum(axis=1)
gender_ratio['MaleRatio'] = round(100 * gender_ratio['M'] / sums , 1)
gender_ratio['MaleRatio'].sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(['occupation'])['age'].agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender']).agg(['mean'])
# users.groupby(['occupation','gender']).mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
#create pivot
gender_ratio = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)

#calc ratio
sums = gender_ratio[['F', 'M']].sum(axis=1)
gender_ratio['MalePercentage'] = round(100 * gender_ratio['M'] / sums , 1)
gender_ratio['FemalePercentage'] = round(100 * gender_ratio['F'] / sums , 1)

# gender_ratio['MalePercentage'].sort_values(ascending=False)
# gender_ratio['FemalePercentage'].sort_values(ascending=False)

print(gender_ratio)
```
